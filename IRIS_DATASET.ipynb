```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('Desktop/MACHINE_LEARNING/MY_DATA_SETS/Iris/Iris.csv')
```

```python
df
```

```python
pd.get_dummies(df['Species'],drop_first=True)
```

```python
specs=pd.get_dummies(df['Species'],drop_first=True)
```

```python
versi=pd.DataFrame(specs['Iris-versicolor'])
virgin=pd.DataFrame(specs['Iris-virginica'])
```

```python
df['versi']=specs['Iris-versicolor']
```

```python
df['virgin']=specs['Iris-virginica']
```

```python
df.drop(['Id','Species'],inplace=True,axis=1)
```

```python
df
```

```python
#STANDARDIZING THE VALUES
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df.drop(['versi','virgin'],axis=1))
```

```python

scaled_features=scaler.transform(df.drop(['versi','virgin'],axis=1))
```

```python
scaled_features
```

```python
df_feat=pd.DataFrame(scaled_features,columns=df.columns[:-2])
```

```python
df_feat.head()
```

```python
#SPLIT DATA INTO TRAIN AND TEST DATA
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X=df_feat
y=df[['versi','virgin']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#IMPORT AND TRAIN MODEL
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#With k = 1
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
predictions=knn.predict(X_test)
```

```python
predictions
```

```python
#CHECKING ACCURACY OF OUR MODEL WITH K = 1
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
#LET US CHECK ERROR RATES FOR DIFFERENT VALUES OF K
```

```python
error_rate=[]

for i in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    predictions_i=knn.predict(X_test)
    error_rate.append(np.mean(predictions_i != y_test))
```

```python
#PLOT THE GRAPH OF ERROR DIFFERENCE
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30),error_rate,color='blue',linestyle='dashed'
        ,marker='o',markerfacecolor='red',markersize=10)
plt.title('Error rate vs K value')
plt.xlabel('K')
plt.ylabel('Error rate')
```

```python
#WE FIND THE THE LOWEST ERROR RATE WE GET IS IN THE RANGE 5-13 (APPROX.)
```
